```python
from scipy import stats

z_scores = stats.zscore(melbourne_data.Rooms.fillna(melbourne_data.Rooms.mean()))
outliers = melbourne_data[z_scores > 3]
```

```python
import matplotlib.pyplot as plt

def boxPlotAll(data):
    plt.boxplot(data)
    plt.xticks(range(1, len(data.columns) + 1), data.columns)
    plt.show()

# Now you can use the function
boxPlotAll(melbourne_data[["Rooms","Price"]])
```

```python
# Remove rows containing nulls
melbourne_data_droprows = melbourne_data.dropna()
melbourne_data_droprows
```

```python
mean = melbourne_data["BuildingArea"].mean()       # calculate the mean for the column
melbourne_data["BuildingArea"].fillna(value=mean)  # replace nulls with the mean
```

```python
median = melbourne_data["BuildingArea"].median()   # calculate the median for the column
melbourne_data["BuildingArea"].fillna(value=median)  # replace nulls with the mean
```

```python
# Compute year
melbourne_data['Date'] = melbourne_data.Date.dt.year
melbourne_data
```
